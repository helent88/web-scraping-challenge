# NASA Mars News

```python
# Visit Nasa news url through splinter module
mars_url = 'https://redplanetscience.com/'

browser.visit(mars_url)

time.sleep(1)

# HTML object
mars_html = browser.html

# Parse HTML object with BeatifulSoup
mars_soup = bs(mars_html, "html.parser")

# Display the result to figure out what you want to scrape
print(mars_soup.prettify())
```

```python
mars_lists = mars_soup.find_all('div', class_ = 'list_text')
mars_lists
```

```python
# Getting all mars news
mars_lists = mars_soup.find_all('div', class_ = 'list_text')

titles_list = []
paragraphs_list = []

# iterate all the news list
for mars_list in mars_lists:
    
    title = mars_list.find('div', class_='content_title').text
    
    paragraph = mars_list.find('div', class_='article_teaser_body').text
    
    # Append both to a list
    titles_list.append(title)
    paragraphs_list.append(paragraph)
    
    # Print all mars news title and paragraph text
    print(f"News title : {title}")
    print(f"Paragraph text : {paragraph}\n")
```

# Mars Hemispheres

```python
# Setup splinter
executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
# use browser to visit the URL 
hemispheres_url = 'https://marshemispheres.com/'

browser.visit(hemispheres_url)
soup = bs(browser.html, 'html.parser')
```

```python
# create list of items that hold all images and all urls
items = soup.find_all('div', class_='item')
print(items)
```

```python
# Create empty list for hemisphere urls
hemisphere_image_urls = []

# Store the main_ul
hemispheres_main_url = 'https://marshemispheres.com/'

for i in items:
    
    img_dict = {}
    # Store title
    title = i.find('h3').text
   
    # Store link that leads to full image website
    partial_img_url = i.find('a', class_='itemLink product-item')['href']
   
    # Visit the link that contains the full image website
    browser.visit(hemispheres_main_url + partial_img_url)
   
    # HTML Object of individual hemisphere information website
    partial_img_html = browser.html
   
    # Parse HTML with Beautiful Soup for every individual hemisphere information website
    soup = bs( partial_img_html, 'html.parser')
   
    # Retrieve full image source
    img_url = hemispheres_main_url + soup.find('img', class_='wide-image')['src']
   
    img_dict["title"] = title
    img_dict["img_url"] = img_url
    print(img_dict["img_url"])
    
    # Append the retreived information into a list of dictionaries
    #hemisphere_image_urls.append({"title" : title, "img_url" : img_url})
    hemisphere_image_urls.append(img_dict)
```
